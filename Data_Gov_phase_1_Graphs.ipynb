```python
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import plotly.express as px
import ipywidgets as widgets
from IPython.display import display, clear_output
```

```python
df=pd.read_csv('cleaned_hotels.csv')
```

```python
df
```

```python
top_countries=pd.DataFrame(df['countyName'].value_counts().head(10))
top_countries.reset_index(inplace=True)
top_countries
```

```python
top_countries_plot=top_countries.plot(kind='bar', x='countyName', y='count', color='green', legend=False)
top_countries_plot.set_title('Top 10 Counties with Most Hotels')
top_countries_plot.set_xlabel('County')
top_countries_plot.set_ylabel('Number of Hotels')
plt.show()
```

```python
top_cities=pd.DataFrame(df['cityName'].value_counts().head(10))
top_cities.reset_index(inplace=True)
top_cities
```

```python
top_cities_plot=top_cities.plot(kind='bar', x='cityName', y='count', color='orange', legend=False)
top_cities_plot.set_title('Top 10 Cities with Most Hotels')
top_cities_plot.set_xlabel('City')
top_cities_plot.set_ylabel('Number of Hotels')
plt.show()
```

```python
df['HotelRating'].value_counts()
```

```python
rating_mapping = {
    'FiveStar': 5,
    'FourStar': 4,
    'ThreeStar': 3,
    'TwoStar': 2,
    'OneStar': 1,
    'A11': 5
}

df['HotelRating'] = df['HotelRating'].map(rating_mapping)
```

```python
df['HotelRating'].value_counts()
```

```python
avg_rating_country=df.groupby('countyName')['HotelRating'].mean().sort_values(ascending=False).head(10)
```

```python
avg_rating_country=pd.DataFrame(avg_rating_country)
avg_rating_country.reset_index(inplace=True)
avg_rating_country
```

```python
avg_rating_country.plot(kind='barh',x='countyName', y='HotelRating',)
plt.title('Top 10 Counties by Average Hotel Rating')
plt.ylabel('County')
plt.xlabel('Average Rating')
plt.show()
```
