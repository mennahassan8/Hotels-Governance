```python
df_cleaned = df.rename(columns={
    ' countyName': 'countyName',
    ' cityName': 'cityName',
    ' HotelName': 'HotelName',
    ' HotelRating': 'HotelRating',
    ' Address': 'Address',
    ' Attractions': 'Attractions',
    ' Description': 'Description',
    ' HotelFacilities': 'HotelFacilities',
    ' PhoneNumber': 'PhoneNumber',
    ' PinCode': 'PinCode',
    ' HotelWebsiteUrl': 'HotelWebsiteUrl'
})
```

```python
df_cleaned=df_cleaned[['countyName', 'cityName', 'HotelName', 'HotelRating',
                 'Address', 'Description',
                 'HotelFacilities', 'PhoneNumber', 'PinCode']]
```

```python
df_cleaned.dropna(inplace=True)
```

```python
df_cleaned.info()
```

```python
df_cleaned = df_cleaned.copy()
schema = pa.DataFrameSchema(
    {
        "countyName": Column(pa.String),
        "cityName": Column(pa.String),
        "HotelName": Column(pa.String),
        "HotelRating": Column(pa.String),
        "Address": Column(pa.String),
        "Description": Column(pa.String, nullable=True),
        "HotelFacilities": Column(pa.String, nullable=True),
        "PhoneNumber": Column(pa.String, nullable=True),
        "PinCode": Column(pa.String, nullable=True)
    },
)
```

```python
try:
    schema.validate(df_cleaned, lazy=True)
except pa.errors.SchemaErrors as exc:
    failure_cases_df = exc.failure_cases
    display(exc.failure_cases)
```
