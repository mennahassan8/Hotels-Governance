```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

def get_primes(min_val, max_val):
    result = []
    for i in range(min_val, max_val):
        if is_prime(i):
            result.append(i)
    return result

def gcd(a, b):
    while b:
        temp = b
        b = a % b
        a = temp
    return a

def mod_inverse(e, phi):
    old_r = e
    r = phi
    old_s = 1
    s = 0
    while r != 0:
        q = old_r // r
        temp_r = old_r - q * r
        old_r = r
        r = temp_r
        temp_s = old_s - q * s
        old_s = s
        s = temp_s
    return old_s % phi

def generate_rsa_keys():
    primes = get_primes(100, 300)
    sample = random.sample(primes, 2)
    p = sample[0]
    q = sample[1]
    n = p * q
    phi = (p - 1) * (q - 1)
    e_candidates = [3, 5, 7, 11, 13, 17, 19, 23]
    e = None
    for c in e_candidates:
        if c < phi and gcd(c, phi) == 1:
            e = c
            break
    d = mod_inverse(e, phi)
    public_key = (e, n)
    private_key = (d, n)
    return public_key, private_key

def rsa_encrypt(text, e, n):
    result = []
    for ch in text:
        ascii_val = ord(ch)
        encrypted = pow(ascii_val, e, n)
        result.append(encrypted)
    return result

def rsa_decrypt(ciphertext, d, n):
    result = []
    for c in ciphertext:
        decrypted = pow(c, d, n)
        char = chr(decrypted)
        result.append(char)
    return ''.join(result)
```

```python
def check_access(users_df, user_name, action, client_ip):

    allowed_users   = ["Menna", "Rozina", "Noura"]
    allowed_actions = ["create", "delete", "get"]

    if user_name in allowed_users and action in allowed_actions and client_ip == "127.0.0.1":
        return True
    else:
        return False

user_name = "Menna"
action = "delete"
client_ip = "127.0.0.1"


if check_access(users_df, user_name, action, client_ip):
    print(f"Access granted for user '{user_name}' to perform action '{action}'.")
else:
    print(f"Access denied for user '{user_name}' to perform action '{action}'.")
```

```python
policy_json = {
    "uid": "1",
    "description": "Menna, Rozina and Noura are allowed to create, delete, get any resources only if the client IP matches 127.0.0.1.",
    "effect": "allow",
    "rules": {
        "subject": [
            {"$.name": {"condition": "Equals", "value": "Max"}},
            {"$.name": {"condition": "Equals", "value": "Nina"}}
        ],
        "resource": {"$.name": {"condition": "RegexMatch", "value": ".*"}},
        "action": [
            {"$.method": {"condition": "Equals", "value": "create"}},
            {"$.method": {"condition": "Equals", "value": "delete"}},
            {"$.method": {"condition": "Equals", "value": "get"}}
        ],
        "context": {"$.ip": {"condition": "CIDR", "value": "127.0.0.1/32"}}
    },
    "targets": {},
    "priority": 0
}

policy = Policy.from_json(policy_json)

storage = MemoryStorage()
storage.add(policy)
pdp = PDP(storage)

request_json = {
    "subject": {"id": "", "attributes": {"name": "Nouran"}},
    "resource": {"id": "", "attributes": {"name": "myrn:example.com:resource:123"}},
    "action": {"id": "", "attributes": {"method": "get"}},
    "context": {"ip": "127.0.0.1"}
}

request = AccessRequest.from_json(request_json)


if pdp.is_allowed(request):
    print("Access granted for user", request_json["subject"]["attributes"]["name"], "to perform action", action)
else:
    print("Access denied for user", request_json["subject"]["attributes"]["name"], "to perform action", action)
```

#### The General Data Protection Regulation, or GDPR, is a European Union (EU) law that governs how organizations within and outside the EU handle the personal data of EU residents. GDPR was adopted by the European Parliament and Council of the EU in 2016 and took effect on 25 May 2018.


### What is the CCPA?
#### The California Consumer Privacy Act (CCPA) is a piece of data privacy legislation that applies to most businesses that process the personal data of California residents. The CCPA gives California residents a certain amount of control over the personal data that businesses collect about them.

```python
def factor_n(n):
    attacked_p = None
    attacked_q = None
    i = 2
    while i <= int(math.sqrt(n)):
        if n % i == 0:
            attacked_p = i
            attacked_q = n // i
            break
        i = i + 1
    return attacked_p, attacked_q

attacked_p = factor_n(n)[0]
attacked_q = factor_n(n)[1]

print(f"  => Found p = {attacked_p}")
print(f"  => Found q = {attacked_q}")
print(f"  => Verify : p × q = {attacked_p * attacked_q} == n = {n} → {attacked_p * attacked_q == n}")

attacked_phi = (attacked_p - 1) * (attacked_q - 1)
print(f"  => φ(n) = ({attacked_p}-1) × ({attacked_q}-1) = {attacked_phi}")


attacked_d = mod_inverse(e, attacked_phi)
print(f"  => Recovered d = {attacked_d}")
print(f"  => Original  d = {d}")
print(f"  => Match        : {attacked_d == d}")

def attack_decrypt(ciphertext, attacked_d, n):
    result = []
    for c in ciphertext:
        decrypted = pow(c, attacked_d, n)
        char = chr(decrypted)
        result.append(char)
    return ''.join(result)

df['Attacked_Decrypted'] = df['Encrypted_HotelName'].apply(
    lambda x: attack_decrypt(x, attacked_d, n)
)


df['Match'] = df['HotelName'] == df['Attacked_Decrypted']

print(df[['HotelName', 'Attacked_Decrypted', 'Match']].head(10).to_string(index=False))

total   = len(df)
matched = df['Match'].sum()

print()
print(f"  Total hotels       : {total}")
print(f"  Correctly attacked : {matched}")
print(f"  Attack success rate: {matched/total*100:.1f}%")
```
